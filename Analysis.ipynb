```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes to the graph
for i in range(1, 6):
    G.add_node(i)

# Add edges to the graph with timestamp for the first time period
G.add_edge(1, 2, start=0, end=1)
G.add_edge(2, 3, start=0, end=1)
G.add_edge(3, 4, start=0, end=1)
G.add_edge(4, 5, start=0, end=1)

# Add edges for the second time period
G.add_edge(1, 2, start=1, end=2)
G.add_edge(2, 3, start=1, end=2)
G.add_edge(3, 5, start=1, end=2)  # Edge from 3 to 4 is removed and edge from 3 to 5 is added

# Add edges for the third time period
G.add_edge(1, 3, start=2, end=3)  # Edge from 1 to 2 is removed and edge from 1 to 3 is added
G.add_edge(3, 5, start=2, end=3)

# Write the graph to a GEXF file
nx.write_gexf(G, "dynamic_graph.gexf")
```
